## Category

```python
plt.figure(figsize=(5,4))
sns.countplot(x=data['Category'])
plt.show()
```

## Sub-Category

```python
data['Sub.Category'].value_counts()
```

```python
data1 = data[data['Category']=='Office Supplies']
data1
```

```python
plt.figure(figsize=(10,6))
sns.set(style='whitegrid')
sns.countplot(x='Category', data=data1, hue='Sub.Category')
plt.title('Office Supplies - Sub category')
plt.legend(loc=2)
plt.show()
```

```python
data2 = data[data['Category']=='Technology']
data2
```

```python
plt.figure(figsize=(10,6))
sns.set(style='ticks', palette='deep')
sns.countplot(x='Category', data=data2, hue='Sub.Category')
plt.title('Technology - Sub category')
plt.legend(loc=1)
plt.show()
```

```python
data3 = data[data['Category']=='Furniture']
data3
```

```python
data['Category'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.set(style='ticks', palette='deep', font_scale=1.0)
sns.countplot(x='Category', data=data3, hue='Sub.Category')
plt.title('Furniture - Sub category')
plt.legend(loc=2)
plt.show()
```

## Ship-Mode

```python
data['Ship.Mode'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Ship.Mode', data = data)
plt.title('Count of Ship-Mode')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set(style='darkgrid', palette='deep', font_scale=1.1)
sns.countplot(x='Category', data=data, hue='Ship.Mode')
plt.title('Category and Ship-Mode')
plt.legend()
plt.show()
```

## Shipping Mode

```python
data['Shipping.Cost'].value_counts()
```

```python
x = data['Ship.Mode'].value_counts().index
x
```

```python
y = data['Ship.Mode'].value_counts().values
y
```

```python
plt.figure(figsize=(5,4))
plt.pie(y, labels=x,startangle= 240, autopct='%0.2f%%')
plt.title('Shipping-Mode')
plt.legend(loc=4)
plt.show()
```

## Profit and Category

```python
plt.figure(figsize=(5,4))
sns.barplot(x='Category', y='Profit', data=data, estimator='sum')
plt.title('Profit by Category')
plt.show()
```

## Shipping Cost and Category

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Ship.Mode', y='Shipping.Cost', data=data, estimator='sum')
plt.title('Shipping cost by Shipping Mode')
plt.show()
```

## Segment

```python
plt.figure(figsize=(5,4))
plt.pie(data['Segment'].value_counts(), labels=data['Segment'].value_counts().index, startangle=60, autopct='%0.2f%%')
plt.title('Segment')
plt.legend(loc=2)
plt.show()
```

## Order Date

```python
data['Year'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.barplot(x='Year', y='Profit', data=data, estimator='sum')
plt.title('Profit by Year')
plt.show()
```

## Region

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Region',y='Sales', data=data, estimator='sum')
plt.title('Sales Performance by Regions')
plt.show()
```

## Market and Profit

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Market', y='Profit', data=data, estimator='sum')
plt.title('Profit by Market')
plt.show()
```
